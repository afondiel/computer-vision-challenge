```python
from datasets import Dataset, DatasetDict, Features, Value, Image

# Define the feature schema
features = Features({
    'index': Value('int32'),
    'image': Image()
})

# Create your data
# data = [
#     {'index': 0, 'image': '/content/blur.png'},
#     {'index': 1, 'image': '/content/low-res.png'},
#     {'index': 2, 'image': '/content/noisy.png'},
#     {'index': 3, 'image': '/content/noisy2.png'},
#     {'index': 4, 'image': '/content/old.png'},
#     {'index': 5, 'image': '/content/old2.png'}
# ]

data = {
    'index': [0, 1, 2, 3, 4, 5],
    'image': ['/content/blur.png', '/content/low-res.png', '/content/noisy.png',
              '/content/noisy2.png', '/content/old.png', '/content/old2.png']
}

print(data)

# Create a Dataset object
dataset = Dataset.from_dict(data, features=features)

# Save the dataset
dataset.save_to_disk('content/dataset')
```
