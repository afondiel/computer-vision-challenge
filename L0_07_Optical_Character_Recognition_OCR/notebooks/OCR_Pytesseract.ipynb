```python
# Import necessary libraries
import pytesseract
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
def display_image_from_url(url):
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    image.load() # debug
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
    return image
```

```python
# Function to perform OCR using PyTesseract
def ocr_pytesseract(image):
    gray_image = image.convert('L')  # Convert image to grayscale
    text = pytesseract.image_to_string(gray_image)
    return text
```

```python
# Example usage
image_url_1 = 'https://bri-wp-images.s3.amazonaws.com/wp-content/uploads/img_LS-001-HandoutA_fillable.jpg'
# image_url_2 = 'https://example.com/sample.jpg'
# image_url_3 = 'https://example.com/sample.jpg'
# image_url_4 = 'https://example.com/sample.jpg'
# image_url_5 = 'https://example.com/sample.jpg'

image = display_image_from_url(image_url_1)
```

```python
image_to_text = ocr_pytesseract(image)
```
